#### New by Groups - byIncomeBracket(), byAgeRange(), and byEnrollmentType

Let's start by creating a an enrollment object.

```
api = Enrollment([6280]).byMonth()
type(api)
```

I want to view enrollment broken down by the predetermined age groups (age_grp_flag) with in the TAF data.

```
df = api.byAgeRange().fetch()
display(df)
```

This default option provides a helpful breakdown of different age groups, but what if I want to view age bands that aren't already within the TAF data?

In this case, we will enter a python dictionary as an arguement, and specify user defined age bands. Simply enter a string as a label, then a list of integers specifying a minimum and maximum range.

```
df = api.byAgeRange({'Teenager': [13,19],'Adult': [20,64],'Senior': [65,125]}).fetch()

display(df)
```

```
print(api.sql())
```

In this example we use 3 different age bands, but this isn't required. The user has the flexibilty to enter as few or as many age bands as they desire.

Continuing with these user defined age bands, let's also break our enrollment query out by income bracket.

```
df = api.byIncomeBracket().fetch()
display(df)
```

#### View Eligible Beneficiaries Who Aren't Enrolled

Let's shift out focus to eligibility. The first step here is to call our original enrollment object as an eligibility object.

```
api = Eligibility(api)
type(api)
```

Now that we have confirmed our enrollment object is converted to an eligibility object, let's view counts for those who are eligible but not enrolled.

```
display(api.notEnrolled().fetch())
```
