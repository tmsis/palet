##### Count enrollees by month that have ICD-10 codes at least once in inpatient or twice in other services.

```
api = Enrollment([6280]).byMonth().having(    \
    Diagnoses.within([                        \
       (ServiceCategory.inpatient, 1),        \
       (ServiceCategory.other_services, 2)],  \
          Diabetes))
```

```
print(api.sql())
```

```
display(api.fetch())
```

##### Creating multiple indicators

```
api_mark = Enrollment([6280]).byMonth()          \
  .mark(                                         \
    Diagnoses.within([                           \
      (ServiceCategory.inpatient, 1),            \
      (ServiceCategory.other_services, 2)],      \
          Diabetes), 'Diabetes')                 \
  .mark(                                         \
     Diagnoses.within([                          \
       (ServiceCategory.inpatient, 1),           \
       (ServiceCategory.other_services, 2)],     \
           AFib), 'AFib')                          
```

```
print(api_mark.sql())
```

```
display(api_mark.fetch())
```

```
api_re = Enrollment([6280]).byMonth()            \
  .byAgeRange()                                  \
  .byEnrollmentType()                            \
  .mark(                                         \
    Diagnoses.within([                           \
      (ServiceCategory.inpatient, 1),            \
      (ServiceCategory.other_services, 2)],      \
          Diabetes), 'Diabetes')                 \
  .mark(                                         \
     Diagnoses.within([                          \
       (ServiceCategory.inpatient, 1),           \
       (ServiceCategory.other_services, 2)],     \
           AFib), 'AFib')
```

```
df = api_re.fetch()
```

```
print(api_re.sql())
```
